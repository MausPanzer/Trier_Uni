```python
import spacy
from spacy import displacy

# 加载spaCy模型，'en_core_web_trf'是内置的BERT模型
nlp = spacy.load("en_core_web_trf")

# 你可以用任何英文句子进行测试
sentence = "The quick brown fox jumps over the lazy dog."

# 使用BERT模型进行分析
doc = nlp(sentence)

# 输出句子的依存关系
print("依存关系（依存树）:")
for token in doc:
    print(f"{token.text:12} {token.dep_:10} {token.head.text}")

# 可视化句法树或依存树
# 这会在浏览器中渲染出图形化的句法树
displacy.render(doc, style="dep", jupyter=True)  # jupyter=True 用于在Jupyter中显示

# 如果你想保存为html文件，也可以这样做
# displacy.serve(doc, style="dep")  # 直接在浏览器展示
# displacy.render(doc, style="dep", page=True)  # 在页面上渲染
```

```python
import spacy
from spacy import displacy

# 加载使用BERT的spaCy模型
nlp = spacy.load("en_core_web_trf")  # 使用带有BERT的transformers模型

# 输入句子
sentence = "The bank of the river is calm, but the bank of the city is crowded."

# 使用BERT模型进行句法分析
doc = nlp(sentence)

# 输出句子的依存关系
print("依存关系（依存树）:")
for token in doc:
    print(f"{token.text:12} {token.dep_:10} {token.head.text}")

# 可视化句法树或依存树
displacy.render(doc, style="dep", jupyter=True)  # 如果是在Jupyter中运行
```

```python
import spacy
from spacy import displacy

# 加载spaCy模型（Word2Vec的默认模型）
nlp_w2v = spacy.load("en_core_web_sm")  # 默认模型是基于Word2Vec的

# 输入句子
sentence = "The bank of the river is calm, but the bank of the city is crowded."

# 使用Word2Vec模型进行句法分析
doc_w2v = nlp_w2v(sentence)

# 输出句子的依存关系
print("依存关系（依存树）使用Word2Vec：")
for token in doc_w2v:
    print(f"{token.text:12} {token.dep_:10} {token.head.text}")

# 可视化句法树或依存树
displacy.render(doc_w2v, style="dep", jupyter=True)  # 如果是在Jupyter中运行
```
