```python
import torch
from allennlp.predictors.predictor import Predictor

# 加载ELMO模型
predictor = Predictor.from_path("https://allennlp.s3.amazonaws.com/models/elmo-crf-ner-2020.02.10.tar.gz")

# 示例文本
text = "Barack Obama was born in Hawaii."

# 进行预测
prediction = predictor.predict(sentence=text)

# 输出命名实体识别结果
print("Tokens:", prediction['words'])
print("Predictions:", prediction['tags'])
```
