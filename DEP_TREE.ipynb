```python
import spacy
from spacy import displacy

# 加载spaCy模型，'en_core_web_trf'是内置的BERT模型
nlp = spacy.load("en_core_web_trf")

# 你可以用任何英文句子进行测试
sentence = "The quick brown fox jumps over the lazy dog."

# 使用BERT模型进行分析
doc = nlp(sentence)

# 输出句子的依存关系
print("依存关系（依存树）:")
for token in doc:
    print(f"{token.text:12} {token.dep_:10} {token.head.text}")

# 可视化句法树或依存树
# 这会在浏览器中渲染出图形化的句法树
displacy.render(doc, style="dep", jupyter=True)  # jupyter=True 用于在Jupyter中显示

# 如果你想保存为html文件，也可以这样做
# displacy.serve(doc, style="dep")  # 直接在浏览器展示
# displacy.render(doc, style="dep", page=True)  # 在页面上渲染
```

```python
# 导入必要的库
import os

def parse_conllu(file_path):
    sentences = []
    sentence = []
    
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            # 忽略注释行（以#开头的行）
            if line.startswith('#'):
                continue
            # 如果是空行，表示句子结束
            elif line.strip() == '':
                if sentence:
                    sentences.append(sentence)
                sentence = []
            else:
                # 将每一行数据拆分为单词的各个属性
                fields = line.strip().split('\t')
                if len(fields) == 10:
                    word_data = {
                        'id': fields[0],
                        'form': fields[1],
                        'lemma': fields[2],
                        'upostag': fields[3],
                        'xpostag': fields[4],
                        'head': fields[6],
                        'dep': fields[7]
                    }
                    sentence.append(word_data)
    
    return sentences

# 使用路径加载文件
file_path = r"C:\Users\13979\Desktop\dataset\ud-treebanks-v2.17\UD_English-EWT\en_ewt-ud-train.conllu"

# 解析文件
parsed_data = parse_conllu(file_path)

# 打印前两个句子的前几行
for sentence in parsed_data[:2]:
    for word in sentence[:5]:  # 显示每个句子的前5个单词
        print(word)
    print("-----")

import spacy
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# 加载 BERT 模型
nlp = spacy.load("en_core_web_trf")
```

```python
import spacy
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# 加载 en_core_web_sm 模型
sm_model = spacy.load("en_core_web_sm")

# 调整标签数量一致性
def adjust_labels_sm(true_labels, predicted_labels):
    if len(true_labels) != len(predicted_labels):
        min_len = min(len(true_labels), len(predicted_labels))  # 取较小长度
        true_labels = true_labels[:min_len]
        predicted_labels = predicted_labels[:min_len]
    return true_labels, predicted_labels

# 解析句子并提取依存关系
def parse_sentence_with_sm_model(sentence):
    doc = sm_model(' '.join([word['form'] for word in sentence]))  # 将句子中的单词拼接成字符串
    predicted_labels = []
    for token in doc:
        predicted_labels.append(token.dep_)  # 提取每个词的依存关系
    return predicted_labels

# 处理前100个句子，确保每个句子的标签一致
true_labels_sm = []
predicted_labels_sm = []

for sentence in parsed_data[:500]:
    true_labels_sentence = [word['dep'] for word in sentence]
    predicted_labels_sentence = parse_sentence_with_sm_model(sentence)
    
    true_labels_sentence, predicted_labels_sentence = adjust_labels_sm(true_labels_sentence, predicted_labels_sentence)
    
    true_labels_sm.extend(true_labels_sentence)
    predicted_labels_sm.extend(predicted_labels_sentence)

# 计算评估指标
precision_sm = precision_score(true_labels_sm, predicted_labels_sm, average='weighted', zero_division=0)
recall_sm = recall_score(true_labels_sm, predicted_labels_sm, average='weighted', zero_division=0)
f1_sm = f1_score(true_labels_sm, predicted_labels_sm, average='weighted', zero_division=0)
accuracy_sm = accuracy_score(true_labels_sm, predicted_labels_sm)

# 打印评估指标
print(f"Precision (SM): {precision_sm}")
print(f"Recall (SM): {recall_sm}")
print(f"F1-Score (SM): {f1_sm}")
print(f"Accuracy (SM): {accuracy_sm}")
```

```python
import spacy
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# 加载 en_core_web_trf 模型
trf_model = spacy.load("en_core_web_trf")

# 调整标签数量一致性
def adjust_labels_trf(true_labels, predicted_labels):
    if len(true_labels) != len(predicted_labels):
        min_len = min(len(true_labels), len(predicted_labels))  # 取较小长度
        true_labels = true_labels[:min_len]
        predicted_labels = predicted_labels[:min_len]
    return true_labels, predicted_labels

# 解析句子并提取依存关系
def parse_sentence_with_trf_model(sentence):
    doc = trf_model(' '.join([word['form'] for word in sentence]))  # 将句子中的单词拼接成字符串
    predicted_labels = []
    for token in doc:
        predicted_labels.append(token.dep_)  # 提取每个词的依存关系
    return predicted_labels

# 处理前100个句子，确保每个句子的标签一致
true_labels_trf = []
predicted_labels_trf = []

for sentence in parsed_data[:500]:
    true_labels_sentence = [word['dep'] for word in sentence]
    predicted_labels_sentence = parse_sentence_with_trf_model(sentence)
    
    true_labels_sentence, predicted_labels_sentence = adjust_labels_trf(true_labels_sentence, predicted_labels_sentence)
    
    true_labels_trf.extend(true_labels_sentence)
    predicted_labels_trf.extend(predicted_labels_sentence)

# 计算评估指标
precision_trf = precision_score(true_labels_trf, predicted_labels_trf, average='weighted', zero_division=0)
recall_trf = recall_score(true_labels_trf, predicted_labels_trf, average='weighted', zero_division=0)
f1_trf = f1_score(true_labels_trf, predicted_labels_trf, average='weighted', zero_division=0)
accuracy_trf = accuracy_score(true_labels_trf, predicted_labels_trf)

# 打印评估指标
print(f"Precision (TRF): {precision_trf}")
print(f"Recall (TRF): {recall_trf}")
print(f"F1-Score (TRF): {f1_trf}")
print(f"Accuracy (TRF): {accuracy_trf}")
```
